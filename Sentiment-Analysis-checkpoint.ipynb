## ETAPE 3: Text Preprocessing

*** cr√©ation d'une fonction qui regroupera les √©tapes cl√©s du traitement du texte. Cette fonction pr√©parera les deux jeux de donn√©es de mani√®re coh√©rente, rendant le texte pr√™t pour les mod√®les de machine learning et deep learning.***

```python
# T√©l√©charger les stopwords si n√©cessaire
nltk.download('stopwords')

# Fonction de pr√©traitement
def preprocess_text(df, text_column):
    STOPWORDS = set(stopwords.words('english'))
    
    # Convertir la colonne en cha√Æne (si n√©cessaire)
    df[text_column] = df[text_column].astype(str)
    
    # √âtendre les contractions
    df[text_column] = df[text_column].apply(contractions.fix)
    
    # Supprimer les chiffres
    df[text_column] = df[text_column].apply(lambda x: ''.join([char for char in x if not char.isdigit()]))
    
    # Convertir en minuscule
    df[text_column] = df[text_column].str.lower()
    
    # Supprimer les balises HTML
    df[text_column] = df[text_column].apply(lambda x: re.sub(r'<.*?>', ' ', x))
    
    # Supprimer les ponctuations
    df[text_column] = df[text_column].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
    
    # Supprimer les caract√®res sp√©ciaux
    df[text_column] = df[text_column].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', x))
    
    # Supprimer les stopwords
    df[text_column] = df[text_column].apply(lambda x: " ".join([word for word in x.split() if word not in STOPWORDS]))
    
    # Supprimer les espaces multiples
    df[text_column] = df[text_column].apply(lambda x: re.sub(r'\s+', ' ', x).strip())
    
    return df
```

```python
# Appliquer la fonction a notre dataframe
df = preprocess_text(df, 'statement')

# Afficher un aper√ßu du DataFrame nettoy√©
print(df.head(10))
```

## 1.  Analyse des Sentiments avec TextBlob

Dans cette section, nous allons utiliser **TextBlob** pour analyser chaque phrase de notre ensemble de donn√©es. TextBlob est un outil puissant de traitement du langage naturel qui calcule la **polarit√©** des phrases. La polarit√© se d√©compose en deux √©l√©ments principaux : **polarit√©** et **subjectivit√©**.

### 1. Polarit√©
La **polarit√©** est une valeur num√©rique qui identifie si une phrase exprime un sentiment **positif** ou **n√©gatif**. Elle varie entre **-1 et 1**, o√π :

- **1** indique un **sentiment tr√®s positif** ou une affirmation bas√©e sur des **faits** plut√¥t que des √©motions.  
- **-1** refl√®te un **sentiment tr√®s n√©gatif** ou une affirmation sans fondement factuel.

En r√©sum√© :  
- Plus la polarit√© est proche de **1**, plus le sentiment est positif.  
- Plus elle est proche de **-1**, plus le sentiment est n√©gatif.

### 2. Subjectivit√©
La **subjectivit√©** mesure dans quelle mesure une phrase refl√®te des **opinions personnelles** ou des **√©motions**, par opposition √† des **faits objectifs**. Ses valeurs oscillent entre **0 et 1** :

- **1** indique une forte subjectivit√©, c‚Äôest-√†-dire une phrase fond√©e sur des **opinions** ou des **√©motions**.  
- **0** sugg√®re une forte objectivit√©, indiquant une phrase bas√©e sur des **faits**.

### R√©sum√©
La polarit√© √©value le **ton √©motionnel** (positif ou n√©gatif) d‚Äôune phrase, tandis que la subjectivit√© d√©termine si l‚Äô√©nonc√© repose sur des **opinions** ou des **faits**. Ces deux m√©triques combin√©es nous permettent d‚Äôobtenir une analyse plus riche des sentiments v√©hicul√©s dans les textes.


```python
# Forme du jeu de donn√©e

df_Textbolb.shape
```

```python
# Tokenisation, Pos_tagging et Lemmatisation

nlp = spacy.load('en_core_web_sm') # Charger le mod√®le spaCy

# D√©finir une fonction pour la Tokenisation, Pos_tagging et Lemmatisation
def lemmatize_text(text):
    doc = nlp(text)  # Traiter le texte avec spaCy
    # Remplacer les tokens par leurs lemmes
    return ' '.join([token.lemma_ if token.lemma_ != '-PRON-' else token.text for token in doc])
```

```python
# Appliquer Tokenisation, Pos_tagging et Lemmatisation sur la colonne "statement"
df_Textbolb['statement'] = df_Textbolb['statement'].apply(lemmatize_text)
```

```python
# Generer la polirit√© pour chaque sentiment

df_Textbolb['polarit√©'] = df_Textbolb['statement'].apply(lambda tweet: TextBlob(tweet).sentiment)
df_Textbolb.head()
```

```python
# Division de la variable polarit√© en polarity et subjectivity

sentiment_series = df_Textbolb['polarit√©'].to_list()
columns = ['polarity', 'subjectivity']
df_bolb = pd.DataFrame(sentiment_series, columns=columns, index=df.index)
df_bolb = df_bolb.round(2)
df_bolb.head()
```

- Les **messages factuels** sont g√©n√©ralement repr√©sent√©s par des **valeurs positives**.  
  - Ils refl√®tent des faits r√©els, souvent soutenus par des preuves concr√®tes.

- En revanche, les **messages √©motionnels** sont souvent repr√©sent√©s par des **valeurs n√©gatives**.  
  - Ces messages expriment un ressenti dict√© principalement par les √©motions, sans preuves tangibles.  
  - Exemple : Une personne peut d√©clarer √™tre en d√©pression non pas parce qu'elle l'est r√©ellement, mais par empathie ou √©motion, s'identifiant √† la souffrance d'autrui.  

Cependant, cette analyse ne garantit pas une v√©rit√© absolue. Elle offre plut√¥t une m√©thode pour √©valuer, en moyenne, la v√©racit√© ou l'origine des messages, tout en tenant compte de la nuance entre √©motion et r√©alit√©.

```python
#Compiler le modele

model.compile(
    loss='categorical_crossentropy',  # 'categorical_crossentropy' pour une classification multi-classes
    optimizer = 'Adam', 
    metrics=['accuracy']  # Mesure la pr√©cision pendant l'entra√Ænement
)
```

##### Pourquoi stratifier‚ÄØ?
La stratification est essentielle lorsque vous travaillez avec des classes d√©s√©quilibr√©es (par exemple, si certaines cat√©gories comme 
***D√©pression*** ou ***Stress*** apparaissent beaucoup plus fr√©quemment que d'autres). Cela garantit que chaque ensemble de donn√©es (entra√Ænement et test) contient une distribution proportionnelle des classes.

```python
# Fonction d'√©valuation du mod√©le

def metrics_score(actual, predicted, classes):
    # G√©n√©rer le rapport de classification sous forme de DataFrame
    report = classification_report(actual, predicted, target_names=classes, output_dict=True)
    report_df = pd.DataFrame(report).transpose()
    
    return report_df

```

```python
# Fonction pour calculer les poids

def compute_weight(df, target):
    values = target.unique()  # Obtenir les valeurs uniques dans target (les classes)
    
    for value in values:
        n_classes = target.nunique()  # Nombre de classes uniques
        n_samples = df.shape[0]  # Nombre total d'√©chantillons
        class_samples = target[target == value].shape[0]  # Nombre d'√©chantillons pour la classe 'value'

        # Calcul du poids pour chaque classe
        class_weight = round(n_samples / (n_classes * class_samples), 2)
        
        print(f"Poids pour la classe '{value}': {class_weight}")
        
compute_weight(df, df['status'])
```

Nous avons constat√© que la distribution des classes est d√©s√©quilibr√©e dans notre jeu de donn√©es. D√®s le d√©part, afin d'am√©liorer les performances de notre mod√®le, nous avons d√©cid√© d'attribuer des poids sp√©cifiques √† chaque classe.

Ces poids sont calcul√©s √† l'aide de la formule suivante :

\[
w_j = \frac{n_{\text{samples}}}{n_{\text{classes}} \times n_{\text{samples}_j}}
\]

### O√π :
- \(w_j\) : Poids attribu√© √† la classe \(j\).
- \(n_{\text{samples}}\) : Nombre total d'√©chantillons (lignes) dans le jeu de donn√©es.
- \(n_{\text{classes}}\) : Nombre total de classes uniques dans la variable cible.
- \(n_{\text{samples}_j}\) : Nombre d'√©chantillons appartenant √† la classe \(j\).

> **Remarque** : Ce r√©sultat peut √©galement √™tre obtenu automatiquement en d√©finissant le param√®tre 
`class_weight='balanced'` dans de nombreux algorithmes d'apprentissage supervis√©.

```python
# Pr√©traitement du texte
class TextPreprocessor:
    def __init__(self):
        self.stopwords = set(stopwords.words('english'))
    
    def preprocess(self, text):
        # Assurez-vous que le texte est une cha√Æne de caract√®res avant d'appliquer les transformations
        if isinstance(text, str):
            # √âtendre les contractions
            text = contractions.fix(text)
            
            # Supprimer les chiffres
            text = ''.join([char for char in text if not char.isdigit()])
            
            # Convertir en minuscule
            text = text.lower()
            
            # Supprimer les balises HTML
            text = re.sub(r'<.*?>', ' ', text)
            
            # Supprimer les ponctuations
            text = text.translate(str.maketrans('', '', string.punctuation))
            
            # Supprimer les caract√®res sp√©ciaux
            text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
            
            # Supprimer les stopwords
            text = " ".join([word for word in text.split() if word not in self.stopwords])
            
            # Supprimer les espaces multiples
            text = re.sub(r'\s+', ' ', text).strip()
        
        return text

# Transformer pour g√©rer les textes invalides
class InvalidTextHandler(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self  # Pas d'op√©ration de "fit" n√©cessaire
    
    def transform(self, X):
        # Assurez-vous que chaque texte dans X est une cha√Æne avant de le traiter
        return [
            "texte_non_valide" if not isinstance(text, str) or text.strip() == "" or re.match(r'^[\W_]+$', text) else text
            for text in X
        ]

# D√©finir votre RandomForest Classifier (ou tout autre mod√®le que vous souhaitez utiliser)
rf_estimator = RandomForestClassifier(n_estimators=100, random_state=42)

# Pipeline complet
pipeline = Pipeline([
    # √âtape de pr√©traitement du texte
    ('text_preprocessing', FunctionTransformer(
        TextPreprocessorFunction(), validate=False)),  # Pr√©traitement du texte
    
    # Gestion des textes invalides
    ('handle_invalid', InvalidTextHandler()),  # Gestion des textes invalides
    
    # Vectorisation avec TF-IDF
    ('vectorizer', TfidfVectorizer(stop_words='english', max_features=5000, ngram_range=(1, 2))),  # Vectorisation TF-IDF
    
    # Classificateur
    ('classifier', rf_estimator)  # Classification avec RandomForest
])


# Entra√Æner le pipeline
pipeline.fit(X_train, y_train)
```
